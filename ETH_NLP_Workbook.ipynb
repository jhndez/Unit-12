```python
# Download/Update the VADER Lexicon
nltk.download("vader_lexicon")

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

#retreive ETH News 
Ethereum_recent_news = newsapi.get_everything(q='Ethereum',
                                          sources=None,
                                          qintitle = 'Ethereum',
                                          exclude_domains = 'techmeme.com,builders.mozilla.community',
                                          from_param='2020-05-11',
                                          to='2020-06-09',
                                          language='en',
                                          sort_by='relevancy')
```

```python
def create_df(news):
    articles = []
    for article in news:
        try:
            title = article["title"]
            text = article["content"]
            date = article["publishedAt"][:10]

            articles.append({
                "title": title,
                "text": text,
                "date": date,
            })
        except AttributeError:
            pass

    return pd.DataFrame(articles)
```

```python
# Create a DataFrame with the news in English
eth_news_df = create_df(Ethereum_recent_news["articles"])
eth_news_df
```

```python
#Sentiment calculation based on compound score
def get_sentiment(score):
  
    result = 0  # Neutral by default
    if score >= 0.05:  # Positive
        result = 1
    elif score <= -0.05:  # Negative
        result = -1

    return result
```

```python
# Sentiment scores dictionaries
title_sent = {
    "title_compound": [],
    "title_pos": [],
    "title_neu": [],
    "title_neg": [],
    "title_sent": [],
}
text_sent = {
    "text_compound": [],
    "text_pos": [],
    "text_neu": [],
    "text_neg": [],
    "text_sent": [],
}

# Get sentiment for the text and the title
for index, row in eth_news_df.iterrows():
    try:
        # Sentiment scoring with VADER
        title_sentiment = analyzer.polarity_scores(row["title"])
        title_sent["title_compound"].append(title_sentiment["compound"])
        title_sent["title_pos"].append(title_sentiment["pos"])
        title_sent["title_neu"].append(title_sentiment["neu"])
        title_sent["title_neg"].append(title_sentiment["neg"])
        title_sent["title_sent"].append(get_sentiment(title_sentiment["compound"]))

        text_sentiment = analyzer.polarity_scores(row["text"])
        text_sent["text_compound"].append(text_sentiment["compound"])
        text_sent["text_pos"].append(text_sentiment["pos"])
        text_sent["text_neu"].append(text_sentiment["neu"])
        text_sent["text_neg"].append(text_sentiment["neg"])
        text_sent["text_sent"].append(get_sentiment(text_sentiment["compound"]))
    except AttributeError:
        pass

# Attaching sentiment columns to the News DataFrame
title_sentiment_df = pd.DataFrame(title_sent)
text_sentiment_df = pd.DataFrame(text_sent)
eth_sentiment_news_df = eth_news_df.join(title_sentiment_df).join(text_sentiment_df)

eth_sentiment_news_df.describe()
```

```python
#function for procuring news from raw API output for BTC recent news 
def create_txt(news):
    articles = []
    for article in news:
        try:
            text = article["content"]
           

            articles.append({
                "text": text,
            
            })
        except AttributeError:
            pass

    return (articles)
```

```python
# getting just str from raw API output fot BTC recent news  
eth_news_txt = create_txt(Ethereum_recent_news["articles"])
eth_values = [d['text'] for d in eth_news_txt]
raw_eth_news_text =''.join(eth_values)

# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

#Define preprocess function for raw_btc_news_text, run function
def process_text(doc):
    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    sw_addons = {'every','daywelcome','charsour'}
    re_clean = regex.sub('', doc)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem if word.lower() not in sw.union(sw_addons)]
    return output

tokenized_eth_news_text = process_text(raw_eth_news_text)

print(tokenized_eth_news_text)
```

```python
# get top 10 word counts
word_counts = Counter(tokenized_eth_news_text)
print(dict(word_counts.most_common(10)))
```

```python
#ngrams 
bigram_counts = Counter(ngrams(tokenized_eth_news_text, n=2))
print(dict(bigram_counts))
```

```python
#func for processing text for btc_news worcloud

def process_text(doc):
    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    sw_addons = {'every','daywelcome','charsour'}
    re_clean = regex.sub('', doc)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem if word.lower() not in sw.union(sw_addons)]
    return ' '.join(output)

# Process text for wordcloud creation
big_str = ''.join(raw_eth_news_text)
input_text_wordcloud_eth = process_text(big_str)
input_text_wordcloud_eth
```
